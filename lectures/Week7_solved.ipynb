**Our World in Data** ([ourworldindata.org](https://ourworldindata.org)): Article on global crime rates. ## Exercise 3.2 — Looking ahead to Week 8

Thinking about the data story:

**Most surprising finding:** Motor Vehicle Theft surged *after* COVID, exceeding pre-pandemic levels — while most crimes declined. This counter-trend deserves explanation.

**Most compelling visualization:** The district crime ratio heatmap — the geographic pattern of over/under-representation is immediately striking and raises questions a general reader can engage with.

**Key takeaway for a non-expert reader:** Crime in San Francisco is not uniformly distributed — it is heavily concentrated in specific neighborhoods, specific hours, and specific categories. Aggregate crime statistics are almost always misleading because they average over wildly skewed distributions.
